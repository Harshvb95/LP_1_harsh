```python
#Importing necessary Libraries
import pandas as pd
from sklearn.cluster import KMeans 
from matplotlib import pyplot as plt 
from sklearn.metrics import silhouette_score
```

```python
#Reading and processing data
df = pd.read_csv(r"/Users/harshavardhan/Desktop/archive/Mall_Customers.csv")
```

```python
x = df.iloc[:, [3,4]]. values
```

```python
x
```

```python
wcss_list = []
```

```python
#The Elbow Curve
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init="k-means++", random_state=55)
    kmeans.fit(x);
    wcss_list.append(kmeans.inertia_)
plt.plot (range (1, 11), wcss_list) 
plt.title("The Elbow Curve") 
plt.xlabel("Number of clusters(k)") 
plt.ylabel ("wcss_list") 
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init="k-means++", random_state=55)
y_pred = kmeans.fit_predict(x)
```

```python
y_pred
```

```python
#Plotting the Cluster
plt.scatter(x[y_pred == 0, 0], x[y_pred == 0, 1], s = 60, c = 'red', label = 'Cluster1')
plt.scatter(x[y_pred == 1, 0], x[y_pred == 1, 1], s = 60, c = 'blue', label = 'Cluster2')
plt.scatter(x[y_pred == 2, 0], x[y_pred == 2, 1], s = 60, c = 'green', label = 'Cluster3')
plt.scatter(x[y_pred == 3, 0], x[y_pred == 3, 1], s = 60, c = 'violet', label = 'Cluster4')
plt.scatter(x[y_pred == 4, 0], x[y_pred == 4, 1], s = 60, c = 'yellow', label = 'Cluster5') 

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 100, c = 'black', label = 'Centroids')

plt.xlabel('Annual Income (k$)') 
plt.ylabel('Spending Score (1-100)') 
plt.legend() 

plt.show()
```

```python
#Silhouette Score
silhouette_score(x, y_pred)
```
