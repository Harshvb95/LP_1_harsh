```python
#Importing Necessary Libraries
import pandas as pd
```

```python
#Reading CSV file
A= pd.read_csv(r"/Users/harshavardhan/Desktop/archive/temperatures.csv")
A
```

```python
#input data
x1 = A[['YEAR']]
x1
```

```python
# output data
y1 = A[['JAN']]
y1
```

```python
#splitting data
from sklearn.model_selection import train_test_split
x1_train,x1_test,y1_train,y1_test= train_test_split(x1,y1,test_size =0.2)
```

```python
#importing linear regression and creating model
from sklearn.linear_model import LinearRegression
model1= LinearRegression()
model1.fit(x1_train,y1_train)
```

```python
print(model1.coef_)
print(model1.intercept_)
```

```python
predict = model1.predict(x1_test)
predict
```

```python
# calculating errors
from sklearn import metrics
metrics.mean_absolute_error(y1_test,predict)
```

```python
MSE = metrics.mean_squared_error(y1_test,predict)
MSE
```

```python
import math
RMSE = math.sqrt(MSE)
RMSE
```

```python
#visualization
import matplotlib.pyplot as plt
plt.title('Temperature plot')
plt.xlabel('YEAR')
plt.ylabel('JAN')
plt.scatter(x1_train,y1_train)
plt.plot(x1_test,predict,color ='red')
```
